```python
class LogisticRegressionGD(object):
    """Logistic Regression Classifier using gradient descent.

    Parameters
    ------------
    eta : float
      Learning rate (between 0.0 and 1.0)
    n_iter : int
      Passes over the training dataset.
    random_state : int
      Random number generator seed for random weight
      initialization.


    Attributes
    -----------
    w_ : 1d-array
      Weights after fitting.
    cost_ : list
      Logistic cost function value in each epoch.

    """
    def __init__(self, eta=0.05, n_iter=100, random_state=1):
        self.eta = eta
        self.n_iter = n_iter
        self.random_state = random_state

    def fit(self, X, y):
        """ Fit training data.

        Parameters
        ----------
        X : {array-like}, shape = [n_samples, n_features]
          Training vectors, where n_samples is the number of samples and
          n_features is the number of features.
        y : array-like, shape = [n_samples]
          Target values.

        Returns
        -------
        self : object

        """
        rgen = np.random.RandomState(self.random_state)
        self.w_ = rgen.normal(loc=0.0, scale=0.01, size=1 + X.shape[1])
        self.cost_ = []

        for i in range(self.n_iter):
            net_input = self.net_input(X)
            output = self.activation(net_input)
            errors = (y - output)
            self.w_[1:] += self.eta * X.T.dot(errors)
            self.w_[0] += self.eta * errors.sum()
            # note that we compute the logistic `cost` now
            # instead of the sum of squared errors cost
            cost = -y.dot(np.log(output)) - ((1 - y).dot(np.log(1 - output)))
            self.cost_.append(cost)
        return self
    
    def net_input(self, X):
        """Calculate net input"""
        return np.dot(X, self.w_[1:]) + self.w_[0]

    def activation(self, z):
        """Compute logistic sigmoid activation"""
        return 1. / (1. + np.exp(-np.clip(z, -250, 250)))

    def predict(self, X):
        """Return class label after unit step"""
        return np.where(self.net_input(X) >= 0.0, 1, 0)
```

```python
from matplotlib.colors import ListedColormap
import matplotlib.pyplot as plt
#定义可以绘制包含测试集在内的决策区域函数plot_decision_regions
def plot_decision_regions(X, y, classifier, test_idx=None, resolution=0.02):

    # setup marker generator and color map
    markers = ('s', 'x', 'o', '^', 'v')
    colors = ('red', 'blue', 'lightgreen', 'gray', 'cyan')
    cmap = ListedColormap(colors[:len(np.unique(y))])

    # plot the decision surface
    x1_min, x1_max = X[:, 0].min() - 1, X[:, 0].max() + 1
    x2_min, x2_max = X[:, 1].min() - 1, X[:, 1].max() + 1
    xx1, xx2 = np.meshgrid(np.arange(x1_min, x1_max, resolution),
                           np.arange(x2_min, x2_max, resolution))
    Z = classifier.predict(np.array([xx1.ravel(), xx2.ravel()]).T)
    Z = Z.reshape(xx1.shape)
    plt.contourf(xx1, xx2, Z, alpha=0.3, cmap=cmap)
    plt.xlim(xx1.min(), xx1.max())
    plt.ylim(xx2.min(), xx2.max())

    for idx, cl in enumerate(np.unique(y)):
        plt.scatter(x=X[y == cl, 0], 
                    y=X[y == cl, 1],
                    alpha=0.8, 
                    c=colors[idx],
                    marker=markers[idx], 
                    label=cl, 
                    edgecolor='black')

    # highlight test samples
    if test_idx:
        # plot all samples
        X_test, y_test = X[test_idx, :], y[test_idx]

        plt.scatter(X_test[:, 0],
                    X_test[:, 1],
                    c='',
                    edgecolor='black',
                    alpha=1.0,
                    linewidth=1,
                    marker='o',
                    s=100, 
                    label='test set')
```

```python
from sklearn.linear_model import LogisticRegression
#按垂直方向（行顺序）堆叠数组构成一个新的数组
X_combined_std = np.vstack ((X_train_std, X_test_std))
#按水平方向（列顺序）堆叠数组构成一个新的数组
y_combined = np. hstack((y_train, y_test))
#调用逻辑回归LogisticRegression()函数
lr = LogisticRegression (C=100.0,random_state=1)
#fit()训练。调用fit(x,y)的方法来训练模型，其中x为数据的属性，y为所属类型。
lr.fit(X_train_std, y_train)
# 绘制分类器的决策区域
plot_decision_regions (X_combined_std,
                       y_combined,
                       classifier=lr, 
                       test_idx=range(218,310)
                       )
plt.xlabel ('PCA1_1 [standardized]')
plt.ylabel ('PCA2_1[standardized]')
plt.legend (loc= 'upper left')
plt.show ()
```

```python
from sklearn.svm import SVC
#设置总体画布大小
fig = plt.figure(figsize=(16,16))
#设置大标题的文字、大小、颜色
fig.suptitle('SVC with precomputed kernel',fontsize=16, color='red')
#循环改变gamma并绘制子图
for i in range(1,5):
    #设定值
    gamma=0.01*10**i
    #模型
    svm = SVC(kernel='rbf', random_state=1, gamma=gamma, C=1.0)
    #训练
    svm.fit(X_train_std, y_train)
    #选择子图绘图
    plt.subplot(2, 2, i)
    #调用绘制决策区域的函数
    plot_decision_regions(X_combined_std, y_combined,
                          classifier=svm, test_idx=range(218, 310))
    #绘图的细节设置
    plt.title('gamma={%.2f}'%gamma)
    plt.xlabel('PCA1_1[standardized]')
    plt.ylabel('PCA2_1[standardized]')
    plt.legend(loc='upper left')
    plt.tight_layout()
#保存图
plt.savefig('支持向量机决策区域main.png', dpi=300)
plt.show()
```

```python
alpha_lasso = 10**np.linspace(-3,1,100)
lasso = Lasso()
coefs_lasso = []

for i in alpha_lasso:
    lasso.set_params(alpha = i)
    lasso.fit(X_train, y_train)
    coefs_lasso.append(lasso.coef_)
    
plt.figure(figsize=(12,10))
ax = plt.gca()
ax.plot(alpha_lasso, coefs_lasso)
ax.set_xscale('log')
plt.axis('tight')# 设置坐标轴时，坐标轴适应数据量 axis
plt.xlabel('alpha')
plt.ylabel('weights: scaled coefficients')
plt.title('Lasso regression coefficients Vs. alpha')
plt.legend(df.drop('y',axis=1, inplace=False).columns)
plt.show()
```
